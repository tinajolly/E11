```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def stats_fun(data1,str1,str2):
    mean1 = np.mean(data1)
    std1 = np.std(data1)
    thstd1 = np.sqrt(len(data1))
    print(str1,str2,"mean =",mean1)
    print(str1,str2,"std =",std1)
    print(str1,str2,"theoretical std =",thstd1)
```

```python
# Thorium
data_ns1 = pd.read_csv('data_ns1.csv')
data_s1 = pd.read_csv('data_s1.csv')
data_s2 = pd.read_csv('data_s2.csv')
data_s3 = pd.read_csv('data_s3.csv')
data_t1 = pd.read_csv('data_t1.csv')
data_t2 = pd.read_csv('data_t2.csv')
data_t3 = pd.read_csv('data_t3.csv')
```

```python
# 2.a
plt.hist(data_ns1["Counts per Interval"],label = "No Shielding",alpha = 0.75)
plt.hist(data_s1["Counts per Interval"],label = "One Al Shield",alpha = 0.75)
plt.hist(data_s2["Counts per Interval"],label = "Two Al Shields",alpha = 0.75)
plt.hist(data_s3["Counts per Interval"],label = "Three Al Shields",alpha = 0.75)
plt.hist(data_t1["Counts per Interval"],label = "One W Shield",alpha = 0.75)
plt.hist(data_t2["Counts per Interval"],label = "Two W Shields",alpha = 0.75)
plt.hist(data_t3["Counts per Interval"],label = "Three W Shields",alpha = 0.75)
plt.legend()
plt.show()
```

```python
stats_fun(data_ns1["Counts per Interval"],'No','shield')
print('')
stats_fun(data_s1["Counts per Interval"],'One','Al shield')
print('')
stats_fun(data_s2["Counts per Interval"],'Two','Al shields')
print('')
stats_fun(data_s3["Counts per Interval"],'Three','Al shields')
print('')
stats_fun(data_t1["Counts per Interval"],'One','W shield')
print('')
stats_fun(data_t2["Counts per Interval"],'Two','W whields')
print('')
stats_fun(data_t3["Counts per Interval"],'Three','W shields')
```

2.a: The theoretical and experimental uncertainties are very close for most of the data, though two of the tungsten tests had larger differences between the two standard deviations. 

2.b: The general trend in the data is that the number of counts decreases as the number of shields increases. When the first shield is placed, the drop is bigger than when the third shield is placed. 

```python
# 3
x = [0,1,2,3]
y1 = [np.sum(data_ns1["Counts per Interval"]),np.sum(data_s1["Counts per Interval"]),np.sum(data_s2["Counts per Interval"]),np.sum(data_s3["Counts per Interval"])]
y2 = [np.sum(data_ns1["Counts per Interval"]),np.sum(data_t1["Counts per Interval"]),np.sum(data_t2["Counts per Interval"]),np.sum(data_t3["Counts per Interval"])]
plt.plot(x,y1)
plt.plot(x,y2)
plt.errorbar(x,y1,[np.std(data_ns1["Counts per Interval"]),np.std(data_s1["Counts per Interval"]),np.std(data_s2["Counts per Interval"]),np.std(data_s3["Counts per Interval"])])
plt.errorbar(x,y2,[np.std(data_ns1["Counts per Interval"]),np.std(data_t1["Counts per Interval"]),np.std(data_t2["Counts per Interval"]),np.std(data_t3["Counts per Interval"])])
plt.yscale('log')
plt.show()
```
