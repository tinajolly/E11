```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data1 = pd.read_csv('Loc1_session1_group2_2024-04-17_14-50-30.csv') # Etcheverry Breezeway
data2 = pd.read_csv('Loc1_session1_group3_2024-04-17_15-13-13.csv') # Mclaughlin Patio
data3 = pd.read_csv('Loc1_session1_group4_2024-04-17_15-34-35.csv') # Between Pimentel and Latimer
data4 = pd.read_csv('Loc1_session1_group5_2024-04-17_15-55-43.csv') # By Haas and Optometry
data5 = pd.read_csv('Loc1_session1_group6_2024-04-17_16-21-01.csv') # Foothill Parking Lot
print(len(data1))
print(len(data2))
print(len(data3))
print(len(data4))
print(len(data5))
```

```python
def stats_fun(data,str1,str2,str3):
    data = data[40:450]
    mean = np.mean(data)
    std = np.std(data)
    thstd = np.sqrt(mean)
    thstd2 = np.sqrt(np.sum(data))/len(data)
    
    plt.plot(data)
    plt.title(str1 + ' vs Time ' + str2)
    plt.xlabel('Time')
    plt.ylabel(str1)
    #plt.savefig('plot ' + str3)
    plt.show()
    
    plt.hist(data)
    plt.title('Distribution of ' + str1 + ' ' + str2)
    plt.xlabel(str1)
    plt.ylabel('Frequency')
    #plt.savefig('hist ' + str3)
    plt.show()
    
    print('Mean =',mean)
    print('Standard Deviation =',std)
    
    if str1 == 'Counts per 2 Seconds':
        exstdstd = thstd/(np.sqrt(2*np.sqrt(len(data))))
        cps = mean/2
        cpsstd = std/2
        print('Theoretical Standard Deviation =',thstd)
        print('Theoretical Standard Deviation 2 =',thstd2)
        print('Uncertainty on the Experimental Standard Deviation =',exstdstd)
        print('Counts per Second =',cps)
        print('Uncertainty on CPS =',cpsstd)
```

```python
stats_fun(data1['Temperature'],'Temperature','on Etcheverry Breezeway','counts etcheverry')
```

```python
stats_fun(data1['counts'],'Counts per 2 Seconds','on Etcheverry Breezeway','counts etcheverry')
```

```python
stats_fun(data2['counts'],'Counts per 2 Seconds','on Mclaughlin Patio','counts mclaughlin')
```

```python
stats_fun(data3['counts'],'Counts per 2 Seconds','Between Pimentel and Latimer','counts pimentel')
```

```python
stats_fun(data4['counts'],'Counts per 2 Seconds','By Haas and Optometry','counts Haas')
```

```python
stats_fun(data5['counts'],'Counts per 2 Seconds','in Foothill Parking Lot','counts foothill')
```
