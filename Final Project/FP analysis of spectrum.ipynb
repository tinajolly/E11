```python
def add_secondary_xaxis(ax, scale_factor):
    secax = ax.secondary_xaxis('top', functions=(lambda x: x * scale_factor, lambda x: x / scale_factor))
    secax.set_xlabel('Energy [keV] (estimate)')
    return secax
```

```python
## Crop data
sd1 = sp_data1.iloc[-1,2:-2]
sd2 = sp_data2.iloc[-1,2:-2]
sd3 = sp_data3.iloc[-1,2:-2]
sd4 = sp_data4.iloc[-1,2:-2]
sd5 = sp_data5.iloc[-1,2:-2]
sd6 = sp_data6.iloc[-1,2:-2]
sd7 = sp_data7.iloc[-1,2:-2]
```

```python
def rebin(data, n=4):
    """
    Rebins the array. n is the divisor. Rebin the data in the grab_data method.
    """
    a = len(data)/n
    new_data = np.zeros((256, 1))
    i = 0
    count = 0
    while i < a:
        temp = sum(data[i:n*(count+1)])
        new_data[count] = temp
        count += 1
        i += n
    return new_data
```

```python
rsp_data1 = rebin(sd1)
rsp_data2 = rebin(sd2)
rsp_data3 = rebin(sd3)
rsp_data4 = rebin(sd4)
rsp_data5 = rebin(sd5)
rsp_data6 = rebin(sd6)
rsp_data7 = rebin(sd7)
```
