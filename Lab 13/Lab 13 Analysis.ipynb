```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data1 = pd.read_csv('Loc1_session1_group1_2024-04-10_16-06-56.csv') # In Jacobs
data2 = pd.read_csv('Loc1_session2_group1_2024-04-10_16-18-39.csv') # By Cory
data3 = pd.read_csv('Loc1_session3_group1_2024-04-10_16-28-32.csv') # By the Bechtel construction
```

```python
def stats_fun(data,str1,str2):
    mean = np.mean(data)
    std = np.std(data)
    thstd = np.sqrt(mean)
    
    plt.plot(data)
    plt.title(str1 + ' vs Time ' + str2)
    plt.xlabel('Time')
    plt.ylabel(str1)
    plt.show()
    
    plt.hist(data)
    plt.title('Distribution of ' + str1)
    plt.xlabel(str1)
    plt.ylabel('Frequency')
    plt.show()
    
    print('Mean =',mean)
    print('Standard Deviation =',std)
    
    if str1 == 'Counts per 2 Seconds':
        thstdstd = thstd/(np.sqrt(2*np.sqrt(len(data))))
        cps = mean/2
        cpsstd = std/2
        print('Theoretical Standard Deviation =',thstd)
        print('Uncertainty on the Theoretical Standard Deviation =',thstdstd)
        print('Counts per Second =',cps)
        print('Uncertainty on CPS =',cpsstd)
        
    if str1 == 'Counts per 5 Seconds':
        thstdstd = thstd/(np.sqrt(2*np.sqrt(len(data))))
        cps = mean/5
        cpsstd = std/5
        print('Theoretical Standard Deviation =',thstd)
        print('Uncertainty on the Theoretical Standard Deviation =',thstdstd)
        print('Counts per Second =',cps)
        print('Uncertainty on CPS =',cpsstd)
```

```python
stats_fun(data1['counts'],'Counts per 2 Seconds','in Jacobs Hall')
```

```python
stats_fun(data1['Temperature'],'Temperature','in Jacobs Hall')
```

```python
stats_fun(data1['Humidity'],'Humidity','in Jacobs Hall')
```

```python
stats_fun(data2['counts'],'Counts per 5 Seconds','by Cory Hall')
```

```python
stats_fun(data2['Temperature'],'Temperature','by Cory Hall')
```

```python
stats_fun(data2['Humidity'],'Humidity','by Cory Hall')
```

```python
stats_fun(data3['counts'],'Counts per 5 Seconds','by Bechtel Hall Construction')
```

```python
stats_fun(data3['Temperature'],'Temperature','by Bechtel Hall Construction')
```

```python
stats_fun(data3['Humidity'],'Humidity','by Bechtel Hall Construction')
```
