```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data_inside = pd.read_csv('inside_data.csv')
data_outside = pd.read_csv('outside_data.csv')
```

```python
plt.plot(data_inside["PM25"],label = "PM25 Inside")
plt.plot(data_outside["PM25"],label = "PM25 Outside")
plt.legend()
plt.show()
```

```python
plt.hist(data_inside["PM25"],label = "PM25 Inside",alpha = 0.75)
plt.hist(data_outside["PM25"],label = "PM25 Outside",alpha = 0.75)
plt.legend()
plt.show()
```

```python
plt.plot(data_inside["PM10"],label = "PM10 Inside")
plt.plot(data_outside["PM10"],label = "PM10 Outside")
plt.legend()
plt.show()
```

```python
plt.hist(data_inside["PM10"],label = "PM10 Inside",alpha = 0.75)
plt.hist(data_outside["PM10"],label = "PM10 Outside",alpha = 0.75)
plt.legend()
plt.show()
```

```python
plt.plot(data_inside["Temperature"][1:],label = "Temperature Inside")
plt.plot(data_outside["Temperature"][1:],label = "Temperature Outside")
plt.legend()
plt.show()
```

```python
plt.hist((data_inside["Pressure"])[1:],label = "Pressure Inside",alpha = 0.75)
plt.hist((data_outside["Pressure"])[1:],label = "Presssure Outside",alpha = 0.75)
plt.legend()
plt.show()
```

3.a: Temperature, humidity, and pressure all had slightly smaller values outside compared to inside while PM10 and PM25 were very similar inside and outside. 

```python
#PM25
PM25_imean = np.mean(data_inside["PM25"])
PM25_imeanstd = np.std(data_inside["PM25"])/(np.sqrt(len(data_inside["PM25"])))
print("PM25 inside mean =",PM25_imean)
print("PM25 inside std =",PM25_imeanstd)

PM25_omean = np.mean(data_outside["PM25"])
PM25_omeanstd = np.std(data_outside["PM25"])/(np.sqrt(len(data_outside["PM25"])))
print("PM25 outside mean =",PM25_omean)
print("PM25 outside std =",PM25_omeanstd)

PM25_diff = np.abs(PM25_imean - PM25_omean)/PM25_imeanstd
print("The difference is",PM25_diff,"standard deviations")
```

```python
#PM10
PM10_imean = np.mean(data_inside["PM10"])
PM10_imeanstd = np.std(data_inside["PM10"])/(np.sqrt(len(data_inside["PM10"])))
print("PM10 inside mean =",PM10_imean)
print("PM10 inside std =",PM10_imeanstd)

PM10_omean = np.mean(data_outside["PM10"])
PM10_omeanstd = np.std(data_outside["PM10"])/(np.sqrt(len(data_outside["PM10"])))
print("PM10 outside mean =",PM10_omean)
print("PM10 outside std =",PM10_omeanstd)

PM10_diff = np.abs(PM10_imean - PM10_omean)/PM10_imeanstd
print("The difference is",PM10_diff,"standard deviations")
```

```python
#Temperature
Temperature_imean = np.mean(data_inside["Temperature"][1:])
Temperature_imeanstd = np.std(data_inside["Temperature"][1:])/(np.sqrt(len(data_inside["Temperature"][1:])))
print("Temperature inside mean =",Temperature_imean)
print("Temperature inside std =",Temperature_imeanstd)

Temperature_omean = np.mean(data_outside["Temperature"][1:])
Temperature_omeanstd = np.std(data_outside["Temperature"][1:])/(np.sqrt(len(data_outside["Temperature"][1:])))
print("Temperature outside mean =",Temperature_omean)
print("Temperature outside std =",Temperature_omeanstd)

Temperature_diff = np.abs(Temperature_imean - Temperature_omean)/Temperature_imeanstd
print("The difference is",Temperature_diff,"standard deviations")
```

```python
#Humidity
Humidity_imean = np.mean(data_inside["Humidity"][1:])
Humidity_imeanstd = np.std(data_inside["Humidity"][1:])/(np.sqrt(len(data_inside["Humidity"][1:])))
print("Humidity inside mean =",Humidity_imean)
print("Humidity inside std =",Humidity_imeanstd)

Humidity_omean = np.mean(data_outside["Humidity"][1:])
Humidity_omeanstd = np.std(data_outside["Humidity"][1:])/(np.sqrt(len(data_outside["Humidity"][1:])))
print("Humidity outside mean =",Humidity_omean)
print("Humidity outside std =",Humidity_omeanstd)

Humidity_diff = np.abs(Humidity_imean - Humidity_omean)/Humidity_imeanstd
print("The difference is",Humidity_diff,"standard deviations")
```

```python
#Pressure
Pressure_imean = np.mean(data_inside["Pressure"][1:])
Pressure_imeanstd = np.std(data_inside["Pressure"][1:])/(np.sqrt(len(data_inside["Pressure"][1:])))
print("Pressure inside mean =",Pressure_imean)
print("Pressure inside std =",Pressure_imeanstd)

Pressure_omean = np.mean(data_outside["Pressure"][1:])
Pressure_omeanstd = np.std(data_outside["Pressure"][1:])/(np.sqrt(len(data_outside["Pressure"][1:])))
print("Pressure outside mean =",Pressure_omean)
print("Pressure outside std =",Pressure_omeanstd)

Pressure_diff = np.abs(Pressure_imean - Pressure_omean)/Pressure_imeanstd
print("The difference is",Pressure_diff,"standard deviations")
```

3.b: Statistically, the means of PM25 and PM10 inside versus outside are not separated by more than three sigma. There were very large differences in the outside means of temprature, humidity, and pressure versus their inside means in terms of their standard deviations. For all three of them, the difference was much greater than 3 sigma.
