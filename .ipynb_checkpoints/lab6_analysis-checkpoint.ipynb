  - NOTE: The variance in your data is not the uncertainty on the mean (central limit theorem)
  - what is your mean uncertainty?
  - How many inside mean standard deviations (1 sigma uncertainty) are your inside and outside measurements separated by?
- If your distributions were not well described by a Gaussian, do you think you took enough data for the central limit theorem to apply?

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_inside = pd.read_csv('inside_data.csv')
data_outside = pd.read_csv('outside_data.csv')
```

```python
def stats_fun(data1,data2,str1,str2,str3):
    mean1 = np.mean(data1)
    std1 = np.std(data1)
    meanstd1 = np.std(data1)/(np.sqrt(len(data1)))
    print(str1,str2,"mean =",mean1)
    print(str1,str2,"std =",std1)
    print(str1,str2,"std (mean) =",meanstd1)

    mean2 = np.mean(data2)
    std2 = np.std(data2)
    meanstd2 = np.std(data2)/(np.sqrt(len(data2)))
    print(str1,str3,"mean =",mean2)
    print(str1,str3,"std =",std2)
    print(str1,str3,"std (mean) =",meanstd2)

    mean_diff = np.abs(mean1 - mean2)/meanstd1
    if str2 == "inside" and str3 == "outside":
        print("The difference is",mean_diff,"inside standard deviations")
    else:
        print("The difference is",mean_diff,"standard deviations")
```

```python
plt.hist(data_inside["PM25"][100:],label = "PM25 Inside",alpha = 0.75)
plt.hist(data_outside["PM25"][100:],label = "PM25 Outside",alpha = 0.75)
plt.legend()
plt.show()
```

```python
#PM25
stats_fun(data_inside["PM25"][100:],data_outside["PM25"][100:],"PM25","inside","outside")
```

```python
plt.hist(data_inside["PM10"][100:],label = "PM10 Inside",alpha = 0.75)
plt.hist(data_outside["PM10"][100:],label = "PM10 Outside",alpha = 0.75)
plt.legend()
plt.show()
```

```python
#PM10
stats_fun(data_inside["PM10"][100:],data_outside["PM10"][100:],"PM10","inside","outside")
```

```python
plt.hist(data_inside["Temperature"][100:],label = "Temperature Inside",alpha = 0.75)
plt.hist(data_outside["Temperature"][100:],label = "Temperature Outside",alpha = 0.75)
plt.legend()
plt.show()
```

```python
#Temperature
stats_fun(data_inside["Temperature"][100:],data_outside["Temperature"][100:],"Temperature","inside","outside")
```

```python
plt.hist(data_inside["Humidity"][100:],label = "Humidity Inside",alpha = 0.75)
plt.hist(data_outside["Humidity"][100:],label = "Humidity Outside",alpha = 0.75)
plt.legend()
plt.show()
```

```python
#Humidity
stats_fun(data_inside["Humidity"][100:],data_outside["Humidity"][100:],"Humidity","inside","outside")
```

```python
plt.hist((data_inside["Pressure"])[100:],label = "Pressure Inside",alpha = 0.75)
plt.hist((data_outside["Pressure"])[100:],label = "Presssure Outside",alpha = 0.75)
plt.legend()
plt.show()
```
