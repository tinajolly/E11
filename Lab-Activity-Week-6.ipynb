```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_inside = pd.read_csv('inside_data.csv')
data_outside = pd.read_csv('outside_data.csv')
```

```python
plt.hist(data_inside["PM25"][100:],label = "PM25 Inside",alpha = 0.75)
plt.hist(data_outside["PM25"][100:],label = "PM25 Outside",alpha = 0.75)
plt.legend()
plt.show()
```

```python
#PM25
PM25_imean = np.mean(data_inside["PM25"][100:])
PM25_istd = np.std(data_inside["PM25"][100:])
PM25_imeanstd = np.std(data_inside["PM25"][100:])/(np.sqrt(len(data_inside["PM25"][100:])))
print("PM25 inside mean =",PM25_imean)
print("PM25 inside std =",PM25_istd)
print("PM25 inside std (mean) =",PM25_imeanstd)

PM25_omean = np.mean(data_outside["PM25"][100:])
PM25_ostd = np.std(data_outside["PM25"][100:])
PM25_omeanstd = np.std(data_outside["PM25"][100:])/(np.sqrt(len(data_outside["PM25"][100:])))
print("PM25 outside mean =",PM25_omean)
print("PM25 inside std =",PM25_ostd)
print("PM25 outside std (mean) =",PM25_omeanstd)

PM25_diff = np.abs(PM25_imean - PM25_omean)/PM25_imeanstd
print("The difference is",PM25_diff,"inside standard deviations")
```
